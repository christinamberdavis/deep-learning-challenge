## Preprocessing

````python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from tensorflow import keras

#Connect to Google Drive to access data resource and export files
from google.colab import drive, files
drive.mount('/content/drive')

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("/content/drive/MyDrive/CoLabs/deep-learning-challenge/Resources/charity_data.csv")
application_df.head()

#Target: IS_SUCCESSFUL
#Features: APPLICATION_TYPE, AFFILIATION, CLASSIFICATION, USE_CASE, ORGANIZATION, INCOME_AMT, SPECIAL_CONSIDERATIONS, ASK_AMT
````

````python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df.drop(columns=['EIN', 'NAME', 'STATUS'], inplace=True)
application_df
````

````python
# Determine the number of unique values in each column.
application_df.nunique()
````

````python
# Look at APPLICATION_TYPE value counts for binning
application_df['APPLICATION_TYPE'].value_counts()
````

````python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#cutoff value = 100
application_types_to_replace = []
for value in application_df['APPLICATION_TYPE']: 
  if value in application_types_to_replace:
    continue
  elif application_df['APPLICATION_TYPE'].value_counts()[value] < 100:
    application_types_to_replace.append(value)

#print(f"Application types to replace: {application_types_to_replace}")

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
````

````python
# Look at CLASSIFICATION value counts for binning
application_df['CLASSIFICATION'].value_counts()
````

````python
# You may find it helpful to look at CLASSIFICATION value counts >1
#create a variables out of the value counts
classification_value_counts = application_df['CLASSIFICATION'].value_counts()
#filter that variable for values with counts >1 and store as new variable
classification_value_counts_gt1 = classification_value_counts[classification_value_counts>1]
#print new variable
classification_value_counts_gt1
````

````python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#cutoff value = 100
classifications_to_replace = []
for value in application_df['CLASSIFICATION']: 
  if value in classifications_to_replace:
    continue
  elif application_df['CLASSIFICATION'].value_counts()[value] < 100:
    classifications_to_replace.append(value)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
````

````python
# Convert categorical data to numeric with `pd.get_dummies`

# non-numeric columns: APPLICATION_TYPE, AFFILIATION, CLASSIFICATION, USE_CASE, ORGANIZATION, INCOME_AMT, SPECIAL_CONSIDERATIONS, ASK_AMT
#create a dataframe with dummy data
optimize_1_df = pd.get_dummies(data=application_df, columns=['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE', 'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS'])
optimize_1_df
````

````python
# Split our preprocessed data into our features and target arrays
y = optimize_1_df['IS_SUCCESSFUL'].values
X = optimize_1_df.drop(['IS_SUCCESSFUL'],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
````

````python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
````

#Optimization 1
## Compile, Train and Evaluate the Model

````python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  50
hidden_nodes_layer2 = 30
hidden_nodes_layer3 = 10
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)
# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

#Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
````

````python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
````

````python
my_callbacks = keras.callbacks.ModelCheckpoint(filepath='/content/drive/MyDrive/CoLabs/deep-learning-challenge/Deep Learning Challenge/Output/saved_model_weights_optimizer1.h5', save_weights_only=True, save_freq=5)

# Train the model
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, callbacks=[my_callbacks])
````

````python
# Evaluate the model using the test data
model_loss_1, model_accuracy_1 = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"\nOptimizer 1 \n Loss: {model_loss_1}, Accuracy: {model_accuracy_1}")
````

````python
# Export our model to HDF5 file
nn.save('/content/drive/MyDrive/CoLabs/deep-learning-challenge/Deep Learning Challenge/Output/AlphabetSoupCharity_Optimized_1.h5')
````
